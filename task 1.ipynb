**DATA CLEANING**

```python
# Need to change the data type of date_added
```

```python
# Converted the data format from "September 24, 2021" to 2021-09-24 for better data analysis and dropped the data_added column
```

```python
df['date_added_in_netflix'] = pd.to_datetime(df['date_added'], format='%B %d, %Y', errors='coerce')
```

```python
df.drop(columns=["date_added"],inplace=True)
```

```python
df.head(2)
```

```python
# Drop the missing values
```

```python
df.dropna(inplace=True)
```

```python
df.head(2)
```

```python
# Dropped the null values
```

```python
df.isnull().sum()
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
df.head(2)
```

```python
# Resetted the index after dropping the null values 
```

```python
df.info()
```

```python
df[['duration_int', 'duration_type']] = df['duration'].str.extract(r'(\d+)\s*(min|Season[s]?)')

df['duration_int'] = pd.to_numeric(df['duration_int'])
```

```python
df.head(2)
```

```python
# created new columns from duration called 'duration_int' and 'duration_type' for better data analyze
# Sepearated the min and seasons
```
